## Â¿QuÃ© variables podrÃ­an estar mÃ¡s relacionadas con el gasto anual de los clientes (Yearly Amount Spent)?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Esta lÃ­nea mÃ¡gica permite que los grÃ¡ficos de Matplotlib se muestren directamente dentro del notebook.
# Aunque en Google Colab suele estar activada por defecto, es buena prÃ¡ctica incluirla.
%matplotlib inline

# Importar el dataset
df = pd.read_csv("../data/Ecommerce Customers.csv")
# Mostramos las primeras 5 filas del DataFrame para tener una idea general del contenido
df.head()

# Mostramos un resumen estadÃ­stico de las variables numÃ©ricas del DataFrame
#df.describe()
plt.figure(figsize=(8, 6))
# Estas variables tienen sentido porque estÃ¡n relacionadas
variables = ['Avg. Session Length','Time on App','Time on Website','Length of Membership','Yearly Amount Spent']
# Paso 3: Usamos un mapa de calor (heatmap) de Seaborn para representar la matriz de correlaciÃ³n.
# - `annot=True` muestra los valores numÃ©ricos dentro de cada celda.
# - `cmap='coolwarm'` es una paleta de colores que va del azul (negativo) al rojo (positivo).
# - `fmt=".2f"` limita los nÃºmeros a dos decimales.
sns.heatmap(df[variables].corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Paso 4: Agregamos un tÃ­tulo al grÃ¡fico
plt.title("Matriz de correlaciÃ³n entre variables")

# Paso 5: Mostramos el grÃ¡fico
plt.show()
```

## Respuesta primera consulta: Segun el anterior diagrama anterior, la variable de membresia ademas y el tiempo en la aplicacion podrian considerarse para el gasto anual

# Â¿Existen diferencias visibles entre los usuarios que usan mÃ¡s la App vs. el Website?

```python
# PRIMER GRAFICO COMPARATIVO DE TIEMPO EN LA APLICACION VS GASTO ANUAL
plt.scatter(x=df['Time on App'],
    y=df['Yearly Amount Spent'], color='blue', alpha=0.5)

# Etiquetamos el eje X para indicar que representa tiempo en la app en condiciones combinadas.
plt.xlabel('Time on App')

# Etiquetamos el eje Y para indicar que representa Gasto Anual
plt.ylabel('Yearly Amount Spent')

# Agregamos un tÃ­tulo descriptivo para que el grÃ¡fico tenga contexto sin necesidad de explicaciÃ³n adicional.
plt.title('Primer Grafico: RelaciÃ³n entre Time on App y Emisiones')

# Agregamos una cuadrÃ­cula de fondo para facilitar la lectura de los valores de los ejes y mejorar la interpretaciÃ³n.
plt.grid(True)

# Finalmente, renderizamos y mostramos el grÃ¡fico en pantalla.
plt.show()

# SEGUNDO GRAFICO COMPARATIVO DE TIEMPO EN LA WEB VS GASTO ANUAL
plt.scatter(x=df['Time on Website'],
    y=df['Yearly Amount Spent'], color='blue', alpha=0.5)

# Etiquetamos el eje X para indicar que representa tiempo en la app en condiciones combinadas.
plt.xlabel('Time on Website')

# Etiquetamos el eje Y para indicar que representa Gasto Anual
plt.ylabel('Yearly Amount Spent')

# Agregamos un tÃ­tulo descriptivo para que el grÃ¡fico tenga contexto sin necesidad de explicaciÃ³n adicional.
plt.title('Primer Grafico: RelaciÃ³n entre Time on Website y Emisiones')

# Agregamos una cuadrÃ­cula de fondo para facilitar la lectura de los valores de los ejes y mejorar la interpretaciÃ³n.
plt.grid(True)

# Finalmente, renderizamos y mostramos el grÃ¡fico en pantalla.
plt.show()
```

## Respuesta: el grafico relacionado con la aplicacion indica que a mas gasto mayor tiempo de uso de la aplicacion, que se puede traducir en compras. A diferencia del segundo grafico donde hay una tendencia nula (por verse horizontal) donde el cliente pasa poco o mucho tiempo, sin mucha variablidad del gasto anual.

# Â¿CÃ³mo se distribuyen los datos para cada variable numÃ©rica? Â¿Existen valores extremos?
