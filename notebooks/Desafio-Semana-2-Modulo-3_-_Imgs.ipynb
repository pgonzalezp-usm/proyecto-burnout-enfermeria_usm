```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from pathlib import Path

#Documents/python/github/proyecto-burnout-enfermeria_usm/data/archive/Brain_Cancer raw MRI data

base_dir = Path("../data/archive/Brain_Cancer raw MRI data/Brain_Cancer")
folders  = ['brain_glioma', 'brain_menin', 'brain_tumor']

dataset = list()
for folder in folders:
    folder_path = base_dir / folder
    
    jpg_files = list(folder_path.glob("*.jpg")) # lista de archivos JPG
    print(f"{folder}: {len(jpg_files)} imágenes")

    for file in jpg_files[:50]:
        # solo cargar primeras 50 imgs por clase
        dataset.append(mpimg.imread(str(file)))
dataset = np.array(dataset)
dataset = dataset[:,:,:,1]     
dataset.shape
plt.imshow(dataset[0,:,:], cmap = 'gray')   
plt.colorbar()
# Verificar el valor máximo en el dataset original
print("maximo sin normalizar ",dataset.max())

# Normalizar los valores de píxel al rango [0, 1], util para analisis de datos
dataset_normalizado = dataset.astype(float) / 255
# Verificar los valores máximo y mínimo después de la normalización
print("max ",dataset_normalizado.max(),";min ", dataset_normalizado.min())
```
